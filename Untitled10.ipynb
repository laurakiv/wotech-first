```python
import numpy as np
import matplotlib.pyplot as plt
import random

num_steps = 10000
position = 0

positions = np.zeros(num_steps + 1)

for step in range(1, num_steps + 1):

    step_direction = np.random.choice([-1, 1])
    position += step_direction
    positions[step] = position

#Plot
plt.figure(figsize=(10, 6))
plt.plot(positions, label='Random Walk', color='blue')
plt.title('1D Simple Random Walk')
plt.xlabel('Number of Steps')
plt.ylabel('Position')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```
